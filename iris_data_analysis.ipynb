# ***Data Visualization***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of petal length vs petal width
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", style="species", s=100)
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```

```python
sns.pairplot(df, hue="species")
```

## Pairplot Insight

The pairplot visualization shows that petal length and petal width provide clear separation between species, with minimal overlap.

In contrast, sepal length and sepal width show significant overlap between versicolor and virginica, making them less effective for classification.

This confirms that petal measurements are the most important features for distinguishing Iris species.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Correlation Heatmap Insight

The correlation heatmap shows that petal length and petal width have the strongest positive correlation among all features.

This indicates that these two features increase together and are highly useful for distinguishing between Iris species.

This further confirms that petal measurements are the most important features for classification tasks.
